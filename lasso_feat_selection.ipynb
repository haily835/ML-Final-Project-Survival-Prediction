```python
import pandas as pd

df = pd.read_csv('./data/processed_mrna_zscore.csv')
df
```

```python
from sklearn.linear_model import Lasso
# Create Lasso regression model
lasso = Lasso(alpha=0.1)  # You can adjust the alpha parameter
X = df.iloc[:, :-2]
y = df.iloc[:, -2]
# Fit the model
lasso.fit(X, y)

# Get feature importance (non-zero coefficients)
selected_features = X.columns[lasso.coef_ != 0]
```

```python
selected_features
```

```python
filepath = f'lasso_feats.txt'
with open(filepath, 'w') as file:
    for item in selected_features:
        file.write("%s\n" % item)
```
