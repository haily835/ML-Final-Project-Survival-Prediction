```python
import pandas as pd
import os

if not os.path.exists('./data'):
    os.makedirs('./data')

prefix_path = './data/brca_metabric'
```

```python
clinical_sample_df = pd.read_csv(f'{prefix_path}/data_clinical_sample.txt', delimiter='\t', skiprows=4)
clinical_sample_df
```

```python
data_clinical_patient = pd.read_csv(f'.{prefix_path}/data_clinical_patient.txt', delimiter='\t')
data_clinical_patient
```

```python
os_months = data_clinical_patient[['PATIENT_ID', 'OS_MONTHS']]
os_months_cleaned = os_months.dropna()
os_months_cleaned
```

```python
os_months_cleaned.describe()
```

```python
methylation_promoters_rrbs_df = pd.read_csv(f'{prefix_path}/data_methylation_promoters_rrbs.txt', delimiter='\t')
methylation_promoters_rrbs_df
```

```python
data_mrna_illumina_microarray_zscores_ref_diploid_samples = pd.read_csv(f'.{prefix_path}/data_mrna_illumina_microarray_zscores_ref_diploid_samples.txt', delimiter='\t')
data_mrna_illumina_microarray_zscores_ref_diploid_samples
```

```python
data_mrna_illumina_microarray = pd.read_csv(f'{prefix_path}/data_mrna_illumina_microarray.txt', delimiter='\t')
data_mrna_illumina_microarray
```

```python
data_mrna_illumina_microarray_transposed = data_mrna_illumina_microarray.transpose()
data_mrna_illumina_microarray_transposed
```

```python
data_mrna_illumina_microarray_transposed = data_mrna_illumina_microarray.transpose()
data_mrna_illumina_microarray_transposed.drop('Entrez_Gene_Id', axis=0, inplace=True)
data_mrna_illumina_microarray_transposed
```

```python
data_mrna_illumina_microarray_transposed.reset_index(inplace=True)
data_mrna_illumina_microarray_transposed
```

```python
data_mrna_illumina_microarray_transposed.columns = data_mrna_illumina_microarray_transposed.iloc[0]
data_mrna_illumina_microarray_transposed = data_mrna_illumina_microarray_transposed[1:].reset_index(drop=True)
```

```python
data_mrna_illumina_microarray_transposed.rename(columns={'Hugo_Symbol': 'PATIENT_ID'}, inplace=True)
data_mrna_illumina_microarray_transposed
```

```python
result_df = pd.merge(data_mrna_illumina_microarray_transposed, os_months_cleaned, on='PATIENT_ID')
result_df.to_csv('./data/merged_mrna_dataset.csv')
```

**Zscore MRNA**

```python
data_mrna_illumina_microarray_zscores_ref_diploid_samples = pd.read_csv(f'.{prefix_path}/data_mrna_illumina_microarray_zscores_ref_diploid_samples.txt', delimiter='\t')
data_mrna_illumina_microarray_zscores_ref_diploid_samples
```

```python
data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed = data_mrna_illumina_microarray_zscores_ref_diploid_samples.transpose()
data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed
```

```python
data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed.drop('Entrez_Gene_Id', axis=0, inplace=True)
data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed.reset_index(inplace=True)

data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed.columns = data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed.iloc[0]
data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed = data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed[1:].reset_index(drop=True)

data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed.rename(columns={'Hugo_Symbol': 'PATIENT_ID'}, inplace=True)

result_zscore_df = pd.merge(data_mrna_illumina_microarray_zscores_ref_diploid_samples_transposed, os_months_cleaned, on='PATIENT_ID')

result_zscore_df.to_csv('./data/merged_mrna_zscore_dataset.csv')
```
